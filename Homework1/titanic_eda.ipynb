**Let's create a new feature.**

```python
def age_category(age):
    """
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    """
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data["Age_category"] = age_categories
```

**Another way is to do it with `apply`.**

```python
data["Age_category"] = data["Age"].apply(age_category)
```

**7. What's the most popular first name among male passengers?**
- Charles
- Thomas
- **William**
- John

```python
def get_first_name(full_name):
    begin_index = full_name.find(".") + 1
    end_index = full_name.find(" ", begin_index + 1)
    if end_index == -1:
        end_index = len(full_name)
    return full_name[begin_index + 1 : end_index]
```

```python
data["First name"] = data["Name"].apply(get_first_name)
```

```python
data["First name"].mode()[0]
```
